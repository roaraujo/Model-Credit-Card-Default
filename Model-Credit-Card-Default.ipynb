# Model Credit Card Default

## Descrição 

Este estudo é baseado no conjunto de dados do UCI Machine Learning Repository (https://archive.ics.uci.edu/ml/datasets/default+of+credit+card+clients) - Irvine, CA: Universidade da Califórnia, Escola de Informação e Ciência da Computação.

Este conjunto de dados contém informações sobre pagamentos inadimplentes, fatores demográficos, dados de crédito, histórico de pagamento e extratos de contas de clientes de cartão de crédito em Taiwan de abril de 2005 a setembro de 2005.

O objetivo é prever se um cliente vai ficar inadimplente no próximo mês ou não. Em outras palavras, é possível prever a inadimplência com um mês de antecedência?

## Informações sobre dataset

Esta pesquisa empregou uma variável binária, pagamento padrão (Sim = 1, Não = 0), como a resposta
variável. Este estudo revisou a literatura e usou as 23 variáveis a seguir como explicativas
variáveis: <br>
X1: Valor do crédito concedido (dólar NT): inclui tanto o crédito individual ao consumidor
e seu crédito familiar (suplementar). <br>
X2: Sexo (1 = masculino; 2 = feminino). <br>
X3: Educação (1 = pós-graduação; 2 = universidade; 3 = ensino médio; 4 = outros). <br>
X4: Estado civil (1 = casado; 2 = solteiro; 3 = outros). <br>
X5: Idade (ano). <br>
X6 - X11: Histórico de pagamentos anteriores. Rastreamos os registros de pagamentos mensais anteriores (de abril
a setembro de 2005) da seguinte forma: X6 = a situação de amortização em setembro de 2005; <br>
X7 = o status de reembolso em agosto de 2005; . . .; <br>
X11 = o status de reembolso em abril de 2005. <br> 
A escala de medição para o estado de reembolso é: -1 = pagar devidamente; 1 = atraso no pagamento de um mês; 2 = atraso no pagamento por dois meses; . . .; 8 = atraso no pagamento por oito meses; 9 =
atraso de pagamento de nove meses e acima. <br>
X12-X17: Valor da fatura (dólar NT). <br> 
X12 = valor da fatura em setembro, 2005; <br>
X13 = valor da fatura em agosto de 2005; . . .; <br>
X17 = valor da fatura em Abril de 2005. <br>
X18-X23: Valor do pagamento anterior (dólar NT). <br>
X18 = valor pago em setembro de 2005;<br>
X19 = valor pago em agosto de 2005; . . .; <br>
X23 = valor pago em abril de 2005 .<br>
Y: pagamento padrão (0 = não padrão; 1 = padrão)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold, learning_curve

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    """Generate a simple plot of the test and training learning curve"""
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```
